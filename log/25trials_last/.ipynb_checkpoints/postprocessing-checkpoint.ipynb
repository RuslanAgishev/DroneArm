```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
def init_fonts(small=12, medium=14, big=18):
    SMALL_SIZE = small
    MEDIUM_SIZE = medium
    BIGGER_SIZE = big

    plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def visualize_flight(trial):
    pose = pd.read_csv('csv/'+str(trial)+'/_vehicle_local_position_0.csv')
    t = ( pose.timestamp - pose.timestamp[0] ) / 10e5
    try:
        sp = pd.read_csv('csv/'+str(trial)+'/_vehicle_local_position_setpoint_0.csv')
        t_sp = ( sp.timestamp - pose.timestamp[0] ) / 10e5
    except:
        print('No setpoint data')
        sp = None

    plt.figure()
    plt.grid()
    plt.title('Flight altitude')
    plt.plot(t, -pose.z, label='altitude')
    if sp is not None: plt.plot(t_sp, -sp.z, label='setpoint')
    plt.xlabel('Time, [sec]')
    plt.ylabel('Z, [m]')
    plt.legend()

    plt.figure()
    plt.grid()
    plt.xlabel('X, [m]')
    plt.ylabel('Y, [m]')
    plt.plot(pose.x, pose.y)
```

```python
init_fonts()
```

```python
# trial could be a number from 1 to 25
visualize_flight(trial=20)
```
